<h3>Data scraping</h3>

```python
from bs4 import BeautifulSoup

import requests
import pandas as pd

import math
```

```python
page = 1
# magic_number, 202

brand_name = []
title = []
color = []
full_price = []
final_price = []

while page != 202:
    base_url = "https://www.endclothing.com/sg/sale?page=" + str(page)
    
    response = requests.get(base_url)
    
    print("status_code: " + str(response.status_code))
    
    while response.status_code != 200:
        response = requests.get(base_url)
    html_content = response.text

    soup = BeautifulSoup(html_content, 'html.parser')
    
    #Get product title
    product_title = soup.find_all(attrs={"data-test-id": "ProductCard__PlpName"})

    for item in product_title:
        title.append(item.get_text())
        
    #Get product color
    product_color = soup.find_all(attrs={"data-test-id": "ProductCard__ProductColor"})

    for item in product_color:
        color.append(item.get_text())
    
    #Get product full price
    product_full_price = soup.find_all(attrs={"data-test-id": "ProductCard__ProductFullPrice"})

    for item in product_full_price:
        full_price.append(item.get_text())
    
    #Get product final price
    product_final_price = soup.find_all(attrs={"data-test-id": "ProductCard__ProductFinalPrice"})

    for item in product_final_price:
        final_price.append(item.get_text())
    
    # #Get brand name
    extracted_links = []
    
    product_brands = soup.find_all(attrs={"data-test-id": "ProductCard__ProductCardSC"})

    for extracted in product_brands:
        extracted_links.append("https://www.endclothing.com/"+ extracted['href'])

    for brand in extracted_links:
        response_brand = requests.get(brand)
        
        while response.status_code != 200:
            response_brand = requests.get(brand)
        html_content_brand = response_brand.text

        soup_soup = BeautifulSoup(html_content_brand, 'html.parser')
        
        brand_brand = soup_soup.find_all(attrs={"property": "name"})
        
        temp_brand = ''
        
        for i in brand_brand:
            temp_brand += i.get_text() + ","
        
        brand_name.append(temp_brand)

    print(f"Done page {page}")
    print(len(final_price))
    print(len(brand_name))
    page += 1
```

<h3>Data Cleaning</h3>

```python
df = pd.read_excel("endclothing_database_unclean.xlsx")

df
```

```python
print(df['Brand Name'].isna().sum())
print(df['Title'].isna().sum())
print(df['Color'].isna().sum())
print(df['Full Price'].isna().sum())
print(df['Final Price'].isna().sum())
```

```python
df = df.drop_duplicates()

df
```

```python
#Get only brand
def get_brand(item):
    items = item.split(",")
    
    for i in items:
        if i != 'Mens' and i != 'Brands':
            return i
        elif i != 'Mens' and 'Brands' not in item:
            return i

brand = [get_brand(str(item)) for item in df['Brand Name']]

# df = df.drop('Brand Name')

df["Brand"] = brand

df
```

```python
print(sorted(df['Brand'].unique()))
```

```python
#remove dollar sign and comma from thousand
full_price_unclean = df['Full Price']
final_price_unclean = df['Final Price']

def clean_currency(item):
    if ',' in item:
        items = item.split(',')
        amount = items[0][-1] + items[1]
    else:
        amount = item[2:]
    return amount

full_price_clean = [clean_currency(x) for x in full_price_unclean]
final_price_clean = [clean_currency(i) for i in final_price_unclean]

df = df.drop(columns=['Full Price', 'Final Price'])

df['Full Price ($)'] = full_price_clean
df['Final Price ($)'] = final_price_clean

df
```

```python
# drop rows with wrong data from scraping

df.drop(df[df['Full Price ($)'] <= df['Final Price ($)']].index, inplace = True)

df
```

```python
for i in sorted(df['Final Price ($)'].unique()):
    if type(i) != str:
        print(i)
```

```python
#calculate discount rate
def round_up(n, decimals=0):
    multiplier = 10 ** decimals
    return math.ceil(n * multiplier) / multiplier

discount = []

full_price_clean = df["Full Price ($)"]
final_price_clean = df["Final Price ($)"]

for full, final in zip(full_price_clean, final_price_clean):
    # ((full-final)/full * 100)
    discounted_price = ((int(full)-int(final))/int(full)) * 100
    discount.append(int(round_up(discounted_price, -1)))
    
df['Discount (%)'] = discount

df
```

```python
#calculate losses
def losses(full, final):
    return int(full)- int(final)

loss = [losses(full, final) for full, final in zip(full_price_clean, final_price_clean)]

df['Losses ($)'] = loss

df
```

```python
df.to_excel("endclothing_database_clean.xlsx" , index=False)
```
